```python
import sys
from PySide2.QtWidgets import QApplication, QDialog, QMainWindow, QMessageBox, QPushButton
from PySide2 import QtCore, QtGui, QtWidgets

class InfoPane(QtWidgets.QWidget):
    """
    """

    def __init__(self):
        super(InfoPane, self).__init__()

        layout = QtWidgets.QVBoxLayout()

        self._dial = QtWidgets.QTextEdit()
        self._dial.setText('test!')
        self._dial.setReadOnly(True)
        layout.addWidget(self._dial)

        self.setLayout(layout)


app = QApplication(sys.argv)

window = InfoPane()
window.show()

app.exec_()
```
